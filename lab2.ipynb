```python
import numpy as np
import matplotlib.pyplot as plt
t=np.arange(0,1,0.001)
f=5
y=np.sin(2*np.pi*f*t)

# plt.plot(t,y,linewidth=2)

sampling_freq = 80
time_interval = 1/sampling_freq

t_samp = np.arange(0,1,time_interval)
y_samp = np.sin(2*np.pi*f*t_samp)

plt.stem(t_samp, y_samp, linefmt='r-')
plt.grid(True)
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude')
plt.title('Continuous Time Signal (sine wave)')
plt.show()





import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate sin(x)
x = np.linspace(0, 2 * np.pi, 1000)  # 0 to 2
y = np.sin(x)

# Step 2: Quantization
levels = 8 # Number of quantization levels
y_min, y_max = -1, 1
step = (y_max - y_min) / levels

# Quantize y
y_q = np.round((y - y_min) / step) * step + y_min

# Step 3: Plot both
plt.plot(x, y, label='Original sin(x)')
plt.step(x, y_q, label='Quantized sin(x)', color='brown')
plt.title('Quantization of sin(x)')
plt.xlabel('x')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()
```
